```python
def intersection(line1, line2):
    (x1, y1) = line1[0]
    (x2, y2) = line1[1]
    (x3, y3) = line2[0]
    (x4, y4) = line2[1]
    try:
        t = ((x1-x3)*(y3-y4) - (y1-y3)*(x3-x4)) / ((x1-x2)*(y3-y4) - (y1-y2)*(x3-x4))
        u = -1*((x1-x2)*(y1-y3) - (y1-y2)*(x1-x3)) / ((x1-x2)*(y3-y4) - (y1-y2)*(x3-x4))
        if(0.0 <= t <= 1.0 and 0.0 <= u <= 1.0):
            px = ((x1*y2 - y1*x2)*(x3-x4) - (x1-x2)*(x3*y4 - y3*x4)) / ((x1-x2)*(y3-y4) - (y1-y2)*(x3-x4))
            py = ((x1*y2 - y1*x2)*(y3-y4) - (y1-y2)*(x3*y4 - y3*x4)) / ((x1-x2)*(y3-y4) - (y1-y2)*(x3-x4))
            return (px,py)
    except:
        pass
    return None

test = intersection(((1,2),(3,2)),((2,1),(2,3)))
print(test)
```

```python
# Transform the list of strings to a list of points
def trace_path(directions):
    path = [(0,0)] # Start at 0,0
    for direction in directions:
        if direction[0] == "U":
            distance = int(direction[1:])
            latest_point = path[-1]
            next_point = (latest_point[0], latest_point[1] + distance)
            path.append(next_point)
        if direction[0] == "D":
            distance = int(direction[1:])
            latest_point = path[-1]
            next_point = (latest_point[0], latest_point[1] - distance)
            path.append(next_point)
        if direction[0] == "R":
            distance = int(direction[1:])
            latest_point = path[-1]
            next_point = (latest_point[0] + distance, latest_point[1])
            path.append(next_point)
        if direction[0] == "L":
            distance = int(direction[1:])
            latest_point = path[-1]
            next_point = (latest_point[0] - distance, latest_point[1])
            path.append(next_point)
    return path

# Create a list of points for each wire
wire_1_path = trace_path(input1)
wire_2_path = trace_path(input2)
```

```python
intersections = []
intersection_steps = []
for wire_1_index in range(0, len(wire_1_path) - 1):
    for wire_2_index in range(0, len(wire_2_path) - 1):
        line1 = (wire_1_path[wire_1_index], wire_1_path[wire_1_index + 1])
        line2 = (wire_2_path[wire_2_index], wire_2_path[wire_2_index + 1])
        intersect = intersection(line1, line2)
        if(intersect != None):
            intersection_steps.append((wire_1_index,wire_2_index))
            intersections.append(intersect)
print(intersections)
```

```python
def taxi_cab_distance(point):
    return abs(point[0]) + abs(point[1])

distances = list(map(taxi_cab_distance, intersections))
print(min(distances))
```

```python
def line_length(point1, point2):
    return taxi_cab_distance(((point1[0] - point2[0]), (point1[1] - point2[1])))

def wire_length(path):
    return sum(map(lambda points : line_length(points[0], points[1]) , zip(path, path[1:])))

step_list = []
for step_index in range(0, len(intersection_steps)):
    step = intersection_steps[step_index]
    intersection = intersections[step_index]
    wire_1 = wire_1_path[0:step[0]]
    wire_2 = wire_2_path[0:step[1]]
    wire_1.append(intersection)
    wire_2.append(intersection)
    step_list.append(wire_length(wire_1) + wire_length(wire_2))

print(step_list)
print(min(step_list))
```
